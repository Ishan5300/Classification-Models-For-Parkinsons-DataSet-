```python
parkinsons_dfs=scale.fit_transform(parkinsons_df[features])
parkinsons_dfs=pd.DataFrame(parkinsons_dfs,columns=features)
parkinsons_dfs['status']=parkinsons_df['status']
```

```python
parkinsons_dfs.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=parkinsons_dfs[features]
y=parkinsons_dfs['status']
```

```python
trainX,testX,trainy,testy=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(0.95)
trainX=pca.fit_transform(trainX)
testX=pca.transform(testX)
```

```python
explained_variance=pca.explained_variance_ratio_
print(explained_variance)
```
